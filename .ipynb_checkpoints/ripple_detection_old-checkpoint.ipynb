```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from tqdm_joblib import tqdm_joblib

import pandas as pd
import load_data
import LFP
from joblib import Parallel, delayed
import h5py

from scipy.signal import find_peaks
import pickle
```

```python
class RippleDataset:
    
    def __init__(self, h5_path):
        self.f = h5py.File(h5_path, 'r')
        self.band_sig = self.f['band_sig']
        self.power = self.f['power']
        self.zscored = self.f['zscored']
        self.peaks = self.f['peaks']

    def get_trace(self, time):
        return {
            'band_sig': self.band_sig[time, :],
            'power': self.power[time, :],
            'zscored': self.zscored[time, :],
        }

    def get_band_sig(self, time):
        if len(time)==0:
            return self.band_sig[:]
        else:
            return self.band_sig[time, :]

    def get_power(self, time):
        if len(time)==0:
            return self.power[:]
        else:
            return self.power[time, :]

    def get_zscored(self, time):
        if len(time)==0:
            return self.zscored[:]
        else:
            return self.zscored[time, :]
    
    def get_peaks(self):
        peaks_grp = self.peaks
        names = sorted(peaks_grp.keys(), key=lambda x: int(x[2:]))
        all_peaks = []
        for ch in names:
            all_peaks.append(peaks_grp[ch][:]) 
        return all_peaks

    def close(self):
        self.f.close()
```

```python
def detect_one_channel(z_ch, peak_thresh, dur, dur_thresh):
    
    candidate_peaks, _ = find_peaks(z_ch, height=peak_thresh, distance=dur)
    true_peaks = []

    for peak_ in candidate_peaks:
        on_ = peak_ - np.argmax(z_ch[:peak_][::-1] < dur_thresh)
        off_ = peak_ + np.argmax(z_ch[peak_:] < dur_thresh)
        if off_ - on_ > dur:
            true_peaks.append(np.array([peak_, on_, off_]))

    return np.stack(true_peaks) if len(true_peaks) else None
```

```python
def get_zscore(date, smooth):
    
    ds = RippleDataset(f"D:/SWR/Session_{date}_ripples_80_180.h5")

    if smooth:
        zscored = ds.get_zscored([])
    else:
        power = ds.get_power([])
        zscored = (power - power.mean(axis=0, keepdims=True)) / power.std(axis=0, keepdims=True)

    return zscored
```

```python
def ripple_detection_(zscored, peak_thresh, dur, dur_thresh):

    n_channels = zscored.shape[1]

    with tqdm_joblib(tqdm(desc="Detecting ripples", total=n_channels)):
        results = Parallel(n_jobs=-1)(
            delayed(detect_one_channel)(zscored[:, ch], peak_thresh, dur, dur_thresh)
            for ch in range(n_channels)
        )

    return results
```

```python
dates = ['031723','032023','032823','033123','040423','040723']

for date in dates:

    out_file = f"D:/SWR/Session_{date}_all_params.h5"
    
    with h5py.File(out_file, "w") as f:
        
        for smooth in [0,1]:
            grp_smooth = f.create_group(f"smooth_{smooth}")
            zscored = get_zscore(date, smooth)

            for peak_thresh in [3,4,5]:
                grp_peak = grp_smooth.create_group(f"peak{peak_thresh}")

                for dur in [20,50,100]:
                    grp_dur = grp_peak.create_group(f"dur{dur}")

                    for dur_thresh in [1,2,3]:
                        print(f"Processing: date={date}, smooth={smooth}, peak={peak_thresh}, dur={dur}, thresh={dur_thresh}")
                        
                        SWR_peaks = ripple_detection_(zscored, peak_thresh, dur, dur_thresh)

                        # group for this parameter set
                        grp_param = grp_dur.create_group(f"thresh{dur_thresh}")
                        
                        # save peaks per channel
                        for ch, arr in enumerate(SWR_peaks):
                            if arr is None or len(arr) == 0:
                                grp_param.create_dataset(f"ch{ch}", data=np.array([]), dtype='float32')
                            else:
                                grp_param.create_dataset(f"ch{ch}", data=arr.astype(np.float32))
```

```python
def identify_ripple(date,SWR_peaks):

    # ds = RippleDataset("D:/SWR/Session_%s_ripples_80_180.h5" % date)
    # SWR_peaks = ds.get_peaks()
    # ds.close()
    
    timestamps, lfpData = load_data.load_LFP(date)
    
    ripple_peak = np.zeros(lfpData.shape)
    ripple_dur = np.zeros(lfpData.shape)
    
    for ii in range(lfpData.shape[1]):
        
        if len(SWR_peaks[ii])>0:
            
            ripple_peak[SWR_peaks[ii][:,0].astype(int),ii]=1
            
            for row in SWR_peaks[ii]:
                
                start = int(row[1])
                end   = int(row[2])
                ripple_dur[start:end, ii] = 1    # mark the entire duration from ripple on to ripple off as 1
    
    channels_detected = ripple_dur.sum(axis=1)
    
    # get ripple dataframe, label each ripple with number of channels detected 
    
    ripple_exist = 1*(channels_detected>0)
    ripple_starts = np.where(np.diff(ripple_exist)==1)[0]
    ripple_stops = np.where(np.diff(ripple_exist)==-1)[0]
    
    
    ofc_nchannels = np.zeros(len(ripple_starts))
    hpc_nchannels = np.zeros(len(ripple_starts))
    
    for ii in range(len(ripple_starts)):
        
        chunk = ripple_dur[ripple_starts[ii]:ripple_stops[ii],:]
        ofc_nchannels[ii] = chunk[:,:96].sum(axis=1).max()
        hpc_nchannels[ii] = chunk[:,96:].sum(axis=1).max()
    
    
    ripple_channel = pd.DataFrame(data={
        'rippleID': np.arange(len(ripple_starts)),
        'on': ripple_starts,
        'off': ripple_stops,
        'dur': ripple_stops - ripple_starts,
        'OFC_channels': ofc_nchannels,
        'HPC_channels': hpc_nchannels
    })

    return ripple_channel
```

```python
smooth_list = [0, 1]
peak_thresh_list = [3, 4, 5]
dur_list = [20, 50, 100]
dur_thresh_list = [1, 2, 3]

results = []        # for summary dataframe
all_ripples = {}    # to store ripple_channel per combination

for smooth, peak_thresh, dur, dur_thresh in itertools.product(
        smooth_list, peak_thresh_list, dur_list, dur_thresh_list):

    # --- Run ripple detection ---
    SWR_peaks = get_detected_SWR(date, smooth, peak_thresh, dur, dur_thresh)

    # --- Summarize per detected ripple ---
    ripple_channel = identify_ripple(date, SWR_peaks)
    
    # --- Compute summary metrics ---
    n = len(ripple_channel)
    dur_mean = ripple_channel.dur.mean() if n > 0 else np.nan
    ofc_undetected = np.mean(ripple_channel.OFC_channels == 0) if n > 0 else np.nan

    # --- Save summary ---
    results.append({
        "smooth": smooth,
        "peak_thresh": peak_thresh,
        "dur": dur,
        "dur_thresh": dur_thresh,
        "n_peaks": n,
        "mean_duration": dur_mean,
        "OFC_undetected": ofc_undetected
    })

    # --- Save ripple_channel object for this parameter set ---
    key = f"s{smooth}_p{peak_thresh}_d{dur}_t{dur_thresh}"
    all_ripples[key] = ripple_channel

df = pd.DataFrame(results)

with open(f"D:/SWR/{date}_SWR_all_ripples_{date}.pkl", "wb") as f:
    pickle.dump(all_ripples, f)
```

```python
def plot_param_heatmaps(df, metric, title=None, cmap="viridis"):
    """
    metric: column name in df ("n_peaks", "mean_duration", "OFC_undetected")
    """
    
    smooth_values = sorted(df["smooth"].unique())
    peak_values = sorted(df["peak_thresh"].unique())
    dur_values = sorted(df["dur"].unique())
    dur_thresh_values = sorted(df["dur_thresh"].unique())

    fig, axes = plt.subplots(
        len(smooth_values), 
        len(peak_values),
        figsize=(14, 8),
        sharex=True, sharey=True
    )

    if title:
        fig.suptitle(title, fontsize=16)

    for i, smooth in enumerate(smooth_values):
        for j, peak in enumerate(peak_values):

            ax = axes[i, j]

            # filter relevant subset
            sub = df[(df.smooth == smooth) & (df.peak_thresh == peak)]

            # create heatmap matrix
            mat = np.zeros((len(dur_values), len(dur_thresh_values)))

            for ii, d in enumerate(dur_values):
                for jj, dt in enumerate(dur_thresh_values):
                    row = sub[(sub.dur == d) & (sub.dur_thresh == dt)][metric]
                    mat[ii, jj] = row.values[0]

            # plot
            im = ax.imshow(mat, origin="lower", cmap=cmap)

            ax.set_title(f"smooth={smooth}, peak={peak}")
            ax.set_xticks(range(len(dur_thresh_values)))
            ax.set_yticks(range(len(dur_values)))
            ax.set_xticklabels(dur_thresh_values)
            ax.set_yticklabels(dur_values)

            if i == len(smooth_values) - 1:
                ax.set_xlabel("dur_thresh")
            if j == 0:
                ax.set_ylabel("dur")

    fig.colorbar(im, ax=axes.ravel().tolist(), shrink=0.8)
    plt.tight_layout()
    plt.show()
```

```python
def get_detected_SWR(date,smooth,peak_thresh, dur, dur_thresh):

    with h5py.File(f"D:/SWR/Session_{date}_all_params.h5", "r") as file:

        peaks = file[f"smooth_{smooth}/peak{peak_thresh}/dur{dur}/thresh{dur_thresh}"]
    
        names = sorted(peaks.keys(), key=lambda x: int(x[2:]))
        SWR_peaks = []
        for ch in names:
            SWR_peaks.append(peaks[ch][:]) 

    
    return SWR_peaks
```

```python
def get_duration(trace, peak_, threshold):
    
    on_ = peak_ - np.argmax(trace[:peak_][::-1]<threshold)
    off_ = peak_ + np.argmax(trace[peak_:]<threshold)

    return off_-on_
```

```python
def get_trial_ripple_rate(peaks,df):
    
    starts = df.start_time.values
    stops = df.stop_time.values
    counts = np.zeros(len(df))
    rates = np.zeros(len(df))
    
    for tt in range(len(df)):
        rel_onset = (peaks>starts[tt]) & (peaks<stops[tt])
        if len(rel_onset)>0:
            counts[tt] = rel_onset.sum()
            rates[tt] = rel_onset.sum()/(stops[tt]-starts[tt])*1000
    
    return counts, rates
```

```python
def get_ITI_ripple_rate(peaks,df):
    
    starts = df.start_time.values[1:]
    stops = df.stop_time.values[:-1]
    counts = np.zeros(len(df))
    rates = np.zeros(len(df))
    
    for tt in range(len(df)-1):
        rel_onset = (peaks>stops[tt]) & (peaks<starts[tt])
        if len(rel_onset)>0:
            counts[tt] = rel_onset.sum()
            rates[tt] = rel_onset.sum()/(starts[tt]-stops[tt])*1000
    
    return counts, rates
```

```python
def get_ripple_raster(peaks,df,window,istrial=0):
    
    ripple_time = np.zeros((len(df),window[1]-window[0]))
    if istrial:
        starts = df.start_time.values
    else:
        starts = df.t_on.values.astype(int)
    
    for tt in range(len(df)):
        rel_onset = (peaks[(peaks> starts[tt]+window[0]) & (peaks<starts[tt]+window[1])]-(starts[tt]+window[0])).astype(int)
    
        if len(rel_onset)>0:
            ripple_time[tt,rel_onset] = 1
    
    smoothed = 1000*np.convolve(ripple_time.mean(axis=0),np.ones(200)/200,mode='same')

    return ripple_time, smoothed
```

```python
from contextlib import contextmanager

@contextmanager
def tqdm_joblib(tqdm_object):
    class TqdmBatchCompletionCallback(joblib.parallel.BatchCompletionCallBack):
        def __call__(self, *args, **kwargs):
            tqdm_object.update(n=self.batch_size)
            return super().__call__(*args, **kwargs)

    old_callback = joblib.parallel.BatchCompletionCallBack
    joblib.parallel.BatchCompletionCallBack = TqdmBatchCompletionCallback
    try:
        yield tqdm_object
    finally:
        joblib.parallel.BatchCompletionCallBack = old_callback
        tqdm_object.close()
```
