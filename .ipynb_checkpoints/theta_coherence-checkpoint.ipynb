```python
def import_LFP(date):

    # get LFP data
    lfpFile = h5py.File('D:/Teleworld/bart_I/raw/spikes/Bart_TeleWorld_v13_%s-spikes.mat' % date, 'r')
    timestamps = np.array(lfpFile['/lfpTimeStamps']).ravel()
    lfpData = lfpFile['/lfpTable']
    
    return timestamps, lfpData
```

```python
def import_behavior(date):
    
    DATA_DIR = 'D:/Teleworld/bart_I/processed/'
    os.chdir(DATA_DIR)

    filename = 'Bart_TeleWorld_v13_%s.nwb' % date
    nwbfile = pynwb.NWBHDF5IO(filename,'r').read()

    return nwbfile
```

```python
def preprocess_LFP(lfp, fs=1000):
    
    # notch filter, remove 60Hz and harmonics, iirnotch is better than butterworth
    
    preprocessed = lfp.copy()
    
    for f0 in [60, 120, 180, 240]:

        b, a = signal.iirnotch(f0,f0/2,fs)
        sos = signal.tf2sos(b,a)
        preprocessed = sosfiltfilt(sos, preprocessed)
    
    return preprocessed
```

```python
def freq_band(preprocessed, band, fs=1000):

    # 1. bandpass, ripple band 150-250 (rat), NHP band 80-180
    
    sos = butter(5, band, btype='bandpass', output='sos', fs=fs)
    band_sig = sosfiltfilt(sos, preprocessed)

    
    # 2. calculate the power and phase of this frequency band

    hilb_sig = hilbert(band_sig)
    power = np.abs(hilb_sig)
    phase = np.angle(hilb_sig)


    return band_sig, power, phase
```

```python
def epoch_aligned_lfp(date,epoch,query,window):
    
    # load behavior
    nwbfile = import_behavior(date)
    timestamps, lfpData = import_LFP(date)
    df = nwbfile.intervals[epoch].to_dataframe().query(query)

    num_ch = lfpData.shape[1]
    aligned_lfp = np.zeros((len(df),num_ch,window[1]-window[0]))  # 1 if happened during ITI, 0 if during a trial
    
    for ii in range(len(df)):

        on = df.iloc[ii].t_on.astype(float)
        idx = (timestamps>on+window[0]) & (timestamps<on+window[1])
        aligned_lfp[ii,:,:] = preprocess_LFP(lfpData[idx,:].T.astype(np.float32))
    
    return df, aligned_lfp
```

```python
def epoch_aligned_band(date,epoch,query,theta,window=[-500,2000]):
    
    # load behavior
    nwbfile = import_behavior(date)
    timestamps, lfpData = import_LFP(date)                                                                                               
    df = nwbfile.intervals[epoch].to_dataframe()
    if epoch!='trials':
        df['istarget'] = df['node']==df['target']
        df = df.query(query)

    aligned_phase = np.zeros((lfpData.shape[1],len(df),window[1]-window[0]))
    aligned_power = np.zeros((lfpData.shape[1],len(df),window[1]-window[0]))

    # band_sig = theta['arr_0'][:,channel]
    power = theta['arr_1']
    phase = theta['arr_2']
    
    for ii in range(len(df)):
    
        on = df.iloc[ii].t_on.astype(float)
        idx = (timestamps>on+window[0]) & (timestamps<on+window[1])
        aligned_phase[:,ii,:]=phase[idx,:].T.astype(np.float32)
        aligned_power[:,ii,:]=power[idx,:].T.astype(np.float32)
    
    return aligned_phase, aligned_power
```

```python
def pick_a_channel(aligned_data):

    total_channels = aligned_data.shape[0]
    OFC_channel_average = aligned_data[:96,:,:].mean(axis=(1,2))
    HPC_channel_average = aligned_data[96:,:,:].mean(axis=(1,2))

    OFC_channel = np.argmax(OFC_channel_average)
    HPC_channel = np.arange(96,total_channels)[np.argmax(HPC_channel_average)]

    return OFC_channel, HPC_channel
```

```python
# parallel processing, spectrogram 

fs = 1000
wavelet = 'cmor1.5-1.0'  # complex Morlet with bandwidth=1.5, center freq=1.0
epochs = ['node_on', 'action_on', 'reward_on']

n_cores = multiprocessing.cpu_count()
print(f" Using {n_cores} CPU cores")

for date in ['032823','033123', '032023', '040423', '040723']:

    theta = np.load(f'D:/Theta/Session {date}, theta.npz', allow_pickle=True)
    all_psd = [[[] for _ in range(3)], [[] for _ in range(3)], [[] for _ in range(2)]]
    all_power = [[[] for _ in range(3)], [[] for _ in range(3)], [[] for _ in range(2)]]

    for ii, epoch in enumerate(epochs):

        if epoch == 'reward_on':
            queries = ['trialerror==1', 'trialerror==0']
        else:
            queries = ['trialerror==1', '(trialerror==0)&(step==0)', '(trialerror==0)&(step>0)']

        df, aligned_lfp = epoch_aligned_lfp(date, epoch, 'trialerror<2', [-1000, 1500])
        df.reset_index(inplace=True, drop=True)

        print('aligned LFP loaded')

        aligned_phase, aligned_power = epoch_aligned_band(date, epoch, 'trialerror<2', theta, window=[-1000, 1500])
        print('aligned band loaded')

        # pick a channel for a given epoch
        OFC_channel, HPC_channel = pick_a_channel(aligned_phase)

        for jj, query in enumerate(queries):
            idx = df.query(query).index
            trialNum = len(idx)
            ofc_lfp = aligned_lfp[idx, OFC_channel, :]
            hpc_lfp = aligned_lfp[idx, HPC_channel, :]

            # ==========================================================
            #  Parallelized Welch PSD computation
            # ==========================================================
            def compute_psd(trial_data):
                f, pxx = signal.welch(trial_data, fs=fs, nperseg=fs,
                                      noverlap=int(fs * 0.9), detrend='linear')
                return pxx

            psd_hpc = Parallel(n_jobs=-1)(
                delayed(compute_psd)(trial) for trial in hpc_lfp
            )
            psd_ofc = Parallel(n_jobs=-1)(
                delayed(compute_psd)(trial) for trial in ofc_lfp
            )

            freqs = f  # same for all
            psd = np.stack([psd_hpc, psd_ofc], axis=0)  # shape (2, trialNum, freqBins)

            # ==========================================================
            #  Parallelized Wavelet Power computation
            # ==========================================================
            frequencies = np.logspace(np.log10(1), np.log10(64), 64) / fs
            scale = pywt.frequency2scale(wavelet, frequencies)

            def compute_wavelet_power(trial_data):
                coeffs, _ = pywt.cwt(trial_data, scale, wavelet, sampling_period=1/fs)
                return np.abs(coeffs) ** 2  # power: (freqs, time)

            hpc_power = Parallel(n_jobs=-1)(
                delayed(compute_wavelet_power)(trial) for trial in hpc_lfp
            )
            ofc_power = Parallel(n_jobs=-1)(
                delayed(compute_wavelet_power)(trial) for trial in ofc_lfp
            )

            power = np.stack([hpc_power, ofc_power], axis=0)  # shape (2, trialNum, freqs, time)

            all_psd[ii][jj] = psd
            all_power[ii][jj] = power

    data = {'psd': all_psd, 'spectrogram': all_power}
    save_fn = f'TFR phase aligned channel Session {date}.pkl'
    with open(save_fn, 'wb') as f:
        pickle.dump(data, f)

    print(f' Finished {date}, saved to {save_fn}')

    plot_spectrogram(all_power,date,0)
    plot_spectrogram(all_power,date,1)
```

```python
def coherogram(sig1, sig2, window = 500, slide=25, fs=1000):
    
    timestamps = np.arange(window, 2500-window+slide, slide)
    
    Cxy = np.zeros((sig1.shape[0],len(timestamps),251))

    for trial in tqdm(range(sig1.shape[0])):

        for ii, tt in enumerate(timestamps):
        
            f_coh, Cxy[trial,ii,:] = coherence(sig1[trial,tt-window:tt+window],hpc_lfp[trial,tt-window:tt+window],fs=fs,nperseg=fs/2,noverlap=fs/2*0.95)

    return f_coh, Cxy
```

```python
def coherogram_single_trial(args):
    sig1, sig2, window, slide, fs = args

    timestamps = np.arange(window, 2500 - window + slide, slide)
    out = np.zeros((len(timestamps), 251))

    for i, tt in enumerate(timestamps):
        f, out[i] = coherence(
            sig1[tt-window : tt+window],
            sig2[tt-window : tt+window],
            fs=fs, nperseg=fs//2, noverlap=int(fs*0.5*0.95)
        )

    return out


def coherogram_parallel(sig1, sig2, window=500, slide=25, fs=1000):

    n_trials = sig1.shape[0]
    timestamps = np.arange(window, 2500 - window + slide, slide)

    # Prepare inputs for each trial
    args_list = [(sig1[i], sig2[i], window, slide, fs)
                 for i in range(n_trials)]

    # Parallel run
    results = Parallel(n_jobs=multiprocessing.cpu_count())(
        delayed(coherogram_single_trial)(args)
        for args in args_list
    )

    # Stack results  shape = (trials, windows, freqs)
    Cxy = np.stack(results, axis=0)

    # return frequency axis also
    f_coh = np.linspace(0, fs/2, 251)

    return f_coh, Cxy
```

```python
# parallel processing, coherogram 

fs = 1000
epochs = ['node_on', 'action_on', 'reward_on']

n_cores = multiprocessing.cpu_count()
print(f" Using {n_cores} CPU cores")

for date in ['032823', '033123', '032023', '040423', '040723']:

    theta = np.load(f'D:/Theta/Session {date}, theta.npz', allow_pickle=True)
    all_coh = [[[] for _ in range(3)], [[] for _ in range(3)], [[] for _ in range(2)]]

    for ii, epoch in enumerate(epochs):

        if epoch == 'reward_on':
            queries = ['trialerror==1', 'trialerror==0']
        else:
            queries = ['trialerror==1', '(trialerror==0)&(step==0)', '(trialerror==0)&(step>0)']

        df, aligned_lfp = epoch_aligned_lfp(date, epoch, 'trialerror<2', [-1000, 1500])
        df.reset_index(inplace=True, drop=True)

        print('aligned LFP loaded')

        aligned_phase, aligned_power = epoch_aligned_band(date, epoch, 'trialerror<2', theta, window=[-1000, 1500])
        print('aligned band loaded')

        # pick a channel for a given epoch
        OFC_channel, HPC_channel = pick_a_channel(aligned_power)

        for jj, query in enumerate(queries):
            
            idx = df.query(query).index
            trialNum = len(idx)
            ofc_lfp = aligned_lfp[idx, OFC_channel, :]
            hpc_lfp = aligned_lfp[idx, HPC_channel, :]

            f_coh, Cxy = coherogram_parallel(ofc_lfp, hpc_lfp)
            all_coh[ii][jj] = Cxy

    data = {'coherence': all_coh}
    save_fn = f'hpc ofc coherence Session {date}.pkl'
    with open(save_fn, 'wb') as f:
        pickle.dump(data, f)

    print(f' Finished {date}, saved to {save_fn}')

    plot_coherogram(all_coh, date)
```
