```python
def plot_sem(time, trace, ax, ii, label):

    colors = sns.color_palette('Purples',5)
    mean_trace = trace.mean(axis=0)
    sem_trace = trace.std(axis=0)/np.sqrt(trace.shape[0])
    ax.plot(time, mean_trace, color=colors[ii], label=label)
    ax.fill_between(time,
                    mean_trace - sem_trace,
                    mean_trace + sem_trace,
                    color=colors[ii],
                    alpha=0.3)
```

```python
def block_progression():
        
    window_size=500 
    model_params = [[] for n in range(len(unitNames))]
    model_pvalues = [[] for n in range(len(unitNames))]
    
    for key in epochs.keys(): 
        
        epoch_spikes = aligned_spikes[key] 
        n_windows = epoch_spikes.shape[1] // window_size 
    
        for n in range(n_windows):
            
            epoch_spikecounts = epoch_spikes[:,n*window_size:(n+1)*window_size,:].sum(axis=1) 
    
            
            for neuron_idx in range(len(unitNames)):
                
                df = pd.DataFrame({
                        "trialNum": trial_df.trial_in_block.values,
                        "FR": epoch_spikecounts[:,neuron_idx]})
                
                model = smf.ols("FR ~ trialNum", data=df).fit()
                model_params[neuron_idx].append(model.params.values[1])
                model_pvalues[neuron_idx].append(model.pvalues.values[1])
    
    model_params = np.array(model_params)
    model_pvalues = np.array(model_pvalues)

    save_path = "D:/Trial/OFC/block_progression_Session_%s.pkl" % date

    # Save both objects together in a dictionary
    with open(save_path, "wb") as f:
        pickle.dump({
            "model_params": model_params,
            "model_pvalues": model_pvalues
        }, f)
```

```python
def load_model_params(region):
        
    model_params = []
    model_pvalues = []
    
    for date in ['031723','032823','033123','032023','040423','040723']:
    
        save_path = f"D:/Trial/{region}/block_progression_Session_{date}.pkl"
    
        # Save both objects together in a dictionary
        with open(save_path, "rb") as f:
            data = pickle.load(f)
    
        model_params.append(data['model_params'])
        model_pvalues.append(data['model_pvalues'])
    
    model_params = np.vstack(model_params)
    model_pvalues = np.vstack(model_pvalues)

    return model_params, model_pvalues
```

```python
def extract_sdf_epochs(
    timestamps,
    neuron_spikes,
    neuron_sdf,
    start_times,
    stop_times,
    warp_len=None
):
    """
    Extract neuron SDF aligned to variable-length epochs.

    Parameters
    ----------
    timestamps : (T,) array
    neuron_sdf : (T, ncell) array
    start_times, stop_times : (ntrial,) arrays
    warp_len : int or None
        If provided, resample each epoch to this length.

    Returns
    -------
    sdf_padded : (ntrial, max_len, ncell) float array (NaN padded)
    sdf_warped : (ntrial, warp_len, ncell) float array or None
    """

    ntrial = len(start_times)
    ncell = neuron_sdf.shape[1]

    aligned = []
    aligned_spikes = []
    lengths = np.zeros(ntrial, dtype=int)

    # extract raw segments
    for i, (t0, t1) in enumerate(zip(start_times, stop_times)):
        idx = (timestamps >= t0) & (timestamps < t1)
        seg = neuron_sdf[idx]
        aligned_spikes.append(neuron_spikes[idx])
        aligned.append(seg)
        lengths[i] = seg.shape[0]

    # pad to max length
    max_len = lengths.max()
    sdf_padded = np.full((ntrial, max_len, ncell), np.nan)

    for i, seg in enumerate(aligned):
        sdf_padded[i, :seg.shape[0]] = seg

    # optional time-warping
    sdf_warped = None
    if warp_len is not None:
        sdf_warped = np.zeros((ntrial, warp_len, ncell))
        for i, seg in enumerate(aligned):
            sdf_warped[i] = resample(seg, warp_len)

    return aligned_spikes, sdf_padded, sdf_warped
```

```python
def plot_trial_iti_sdf(
    warped_trial,
    iti_sdf,
    trial_df,
    neuron_idx,
    n_blocks=7
):
    """
    Plot mean SDF across trials within each block (1n_blocks layout).
    """

    plt.close('all')
    colors = sns.color_palette('Purples_r', n_blocks)

    fig, axes = plt.subplots(
        1, n_blocks,
        figsize=(2.2 * n_blocks, 4),
        sharey=True
    )

    trial_len = warped_trial.shape[1]
    iti_len   = iti_sdf.shape[1]

    for b in range(n_blocks):
        ax = axes[b]
        trial_idx = trial_df.trial_in_block == b

        # trial
        ax.plot(
            np.arange(trial_len),
            np.nanmean(warped_trial[trial_idx, :, neuron_idx], axis=0),
            color='k'
        )

        # ITI
        ax.plot(
            np.arange(trial_len, trial_len + iti_len),
            np.nanmean(iti_sdf[trial_idx, :, neuron_idx], axis=0),
            color='k'
        )

        ax.set_title(f"Block {b}", fontsize=9)

    # shared labels
    axes[0].set_ylabel("Firing rate (SDF)")
    fig.supxlabel("Time (trial  ITI)", y=0.05)

    sns.despine()
    fig.suptitle(
        f"Session {date} | Unit {unitNames.iloc[neuron_idx].unitNumber}",
        y=1.02,
        fontsize=12
    )

    plt.tight_layout()
    plt.savefig(
        f"D:/Trial/HPC/within_block_Session_{date}_neuron_{unitNames.iloc[neuron_idx].unitNumber}",
        dpi=300,
        bbox_inches="tight"
    )
```

```python
def plot_trial_iti_goal_sdf(
    warped_trial,
    iti_sdf,
    trial_df,
    neuron_idx
):
    """
    Plot mean SDF across trials within each block (1n_blocks layout).
    """

    plt.close('all')
    colors = sns.color_palette('Set1', 4)

    fig, axes = plt.subplots(
        1, 2,
        figsize=(14, 8),
        sharey=True
    )

    trial_len = warped_trial.shape[1]
    iti_len   = iti_sdf.shape[1]
    titles = ['Navigation','Flash']

    for b in range(2):
        
        ax = axes[b]

        for gg, goal in enumerate([1,7,8,14]):

            if b==0:
                trial_idx = (trial_df.trialerror == b) & (trial_df.target==goal) \
                            & ~np.isin(trial_df.start.values,[1,7,8,14])
            else:
                trial_idx = (trial_df.trialerror == b) & (trial_df.target==goal)
                
    
            # trial
            ax.plot(
                np.arange(trial_len),
                np.nanmean(warped_trial[trial_idx, :, neuron_idx], axis=0),
                color=colors[gg]
            )
    
            # ITI
            ax.plot(
                np.arange(trial_len, trial_len + iti_len),
                np.nanmean(iti_sdf[trial_idx, :, neuron_idx], axis=0),
                color=colors[gg]
            )
    
        ax.set_title(titles[b], fontsize=16)

    # shared labels
    axes[0].set_ylabel("Firing rate (SDF)")
    fig.supxlabel("Time (trial  ITI)", y=0.05, fontsize=18)

    sns.despine()
    fig.suptitle(
        f"Session {date} | Unit {unitNames.iloc[neuron_idx].unitNumber}",
        y=1.02,
        fontsize=12
    )

    plt.tight_layout()
    plt.savefig(
        f"D:/Trial/HPC/Goal_Session_{date}_neuron_{unitNames.iloc[neuron_idx].unitNumber}",
        dpi=300,
        bbox_inches="tight"
    )
```
