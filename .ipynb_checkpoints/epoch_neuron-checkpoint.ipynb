```python
def plot_sem(time, trace, ax, ii, label):

    colors = sns.color_palette('Set1',3)
    mean_trace = trace.mean(axis=0)
    sem_trace = trace.std(axis=0)/np.sqrt(trace.shape[0])
    ax.plot(time, mean_trace, color=colors[ii], label=label)
    ax.fill_between(time,
                    mean_trace - sem_trace,
                    mean_trace + sem_trace,
                    color=colors[ii],
                    alpha=0.3)
```

```python
def raster_sdf_aligned(
    timestamps,
    neuron_spikes,
    neuron_sdf,
    event_times,
    window
):
    """
    event_times: array of event timestamps (ms)
    window: (t_start, t_end) relative to event
    """
    ncell = neuron_spikes.shape[1]
    aligned_spikes = np.zeros((len(event_times),window[1]-window[0],ncell), dtype=int)
    aligned_sdf = np.zeros((len(event_times),window[1]-window[0],ncell), dtype=int)

    for i, t0 in enumerate(event_times):
        t_start = t0 + window[0]
        t_end   = t0 + window[1]

        idx = (timestamps >= t_start) & (timestamps < t_end)
        aligned_spikes[i] = neuron_spikes[idx]
        aligned_sdf[i] = neuron_sdf[idx]

    return aligned_spikes, aligned_sdf
```

```python
def epoch_permutation():
    
    trial_spikes = data['trial_spikes']
    n_neurons = trial_spikes[0].shape[1]
    
    trial_length = -trial_df.start_time.values+trial_df.stop_time.values
    node_ons = node_df.t_on.values-trial_df.start_time.values
    
    n_perm = 1000
    node_FR = np.zeros((len(trial_df),n_neurons))
    null_FR = np.zeros((n_perm,len(trial_df),n_neurons))
    
    for trial in range(len(trial_df)):
        
        valid = trial_spikes[trial]
        first_node_time = np.arange(node_ons[trial],node_ons[trial]+500).astype(int)
        node_FR[trial] = valid[first_node_time,:].sum(axis=0)
        
        trial_time = np.arange(trial_length[trial])
        rest = trial_time[~np.isin(trial_time, np.arange(node_ons[trial]-500,node_ons[trial]+500)) & (trial_time < (trial_length[trial]-500))]
        
        for n in range(n_perm):
            null_idx = int(np.random.choice(rest))
            null_time = np.arange(null_idx,null_idx+500).astype(int)
            null_FR[n,trial,:] = valid[null_time,:].sum(axis=0)
    
    save_path = "D:/Epoch/HPC/First_node_permutation_Session_%s.pkl" % date
    
    # Save both objects together in a dictionary
    with open(save_path, "wb") as f:
        pickle.dump({
            "null": null_FR,
            "node": node_FR
        }, f)
```

```python
def get_concat_sdf():
    
    window_size = 300  # rows per window
    
    concat_sdf_windowed = []
    
    for key in aligned_sdf.keys():
        # mean across trials
        sdf_mean = aligned_sdf[key].mean(axis=0)  # shape: (time, n_neurons)
        
        # compute number of full windows
        n_windows = sdf_mean.shape[0] // window_size
        
        # take full windows and compute mean along each window
        sdf_windowed = sdf_mean[:n_windows*window_size].reshape(n_windows, window_size, -1).mean(axis=1)
        
        concat_sdf_windowed.append(sdf_windowed)
    
    # stack all keys
    concat_sdf_windowed = np.vstack(concat_sdf_windowed)
    
    concat_sdf = np.vstack([aligned_sdf[key].mean(axis=0) for key in aligned_spikes.keys()])
    
    row_mean = concat_sdf.mean(axis=0, keepdims=True)
    row_std = concat_sdf.std(axis=0, keepdims=True)
    
    concat_sdf_norm = (concat_sdf - row_mean) / row_std
    
    
    # concat_sdf: (630, 90)
    # row_min = concat_sdf.min(axis=0, keepdims=True)
    # row_max = concat_sdf.max(axis=0, keepdims=True)
    
    # concat_sdf_norm = (concat_sdf - row_min) / (row_max - row_min)
    
    sort_idx = np.argsort(np.argmax(concat_sdf_windowed,axis=0))

    return concat_sdf_norm, sort_idx
```

```python
def plot_heatmap(concat_sdf_norm,sort_idx,region):
    # Example: durations for relative widths
    durations = np.array([
        epoch_windows[e][1] - epoch_windows[e][0]
        for e in epoch_windows.keys()
    ])
    width_ratios = durations / durations.min()  # relative widths
    
    # Compute shared vmin/vmax across all data
    vmin = -2 #np.nanmin(concat_sdf_norm)
    vmax = 2 #np.nanmax(concat_sdf_norm)
    
    plt.close('all')
    fig, axes = plt.subplots(
        1, len(epoch_windows),
        figsize=(18, 6),
        sharey=True,
        gridspec_kw={"width_ratios": width_ratios}
    )
    
    time_idx = 0
    for jj, key in enumerate(aligned_sdf.keys()):
        dur = np.diff(epoch_windows[key])[0]
        
        sns.heatmap(
            concat_sdf_norm.T[sort_idx, time_idx:time_idx+dur],
            ax=axes[jj],
            cmap='bwr',
            cbar=(jj==len(epoch_windows)-1),  # show colorbar only on last
            vmin=vmin,
            vmax=vmax
        )
        
        # mark event
        axes[jj].axvline(-epoch_windows[key][0], color='w')
        axes[jj].set_title(key)
        time_idx += dur
    
    fig.text(0.02, 0.5, "Neurons", va='center', rotation='vertical')
    plt.tight_layout(rect=[0,0,1,0.9])
    plt.suptitle(f'{region}, trial progression heatmap std')
    plt.savefig(f'D:/Epoch/{region}/trial progression heatmap std')
```
